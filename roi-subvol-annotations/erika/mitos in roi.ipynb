```python
import pandas as pd
import glob
```

```python
def get_stuff(file):
    split_file = file.split("/")
    get_file = split_file[1]
    find_user = get_file.split("_")
    print(find_user)
    user = find_user[1]
    
    file_points = []
    
    out_data = []
    roi_file = "prec_rec/prod_roi_list.txt"
    with open(roi_file) as f:
        roi_info = f.readlines()
    for line in roi_info:
        line = line.strip()
        split_line = line.split("_")
        roi_name = split_line[0]
        roi_start = split_line[1]
        print(roi_name, roi_start)
        
        roi_loc = roi_start.split(", ")
        print(roi_loc)
        
        roi_x_min = int(roi_loc[0])
        roi_y_min = int(roi_loc[1])
        roi_z_min = int(roi_loc[2])
        print("roi start: ",roi_x_min, roi_y_min, roi_z_min)

        roi_x_max = (roi_x_min + 500)
        roi_y_max = (roi_y_min + 500)
        roi_z_max = (roi_z_min + 500)
    
    
        print("start mitos from: ", file)
        df = pd.read_csv(file)
    
       
        roi_mitos = []
        
        mito_locs = df["location"].tolist()
        for loc in mito_locs:
#             print("mito loc: ", loc)
#             fix_loc = loc[1:-1]
            fix_loc = loc
#             print(fix_loc)
            split_loc = fix_loc.split("_")
            mito_x = int(split_loc[0])
            mito_y = int(split_loc[1])
            mito_z = int(split_loc[2])

            status = []

            if mito_x >= roi_x_min and mito_x <= roi_x_max:
                x_stat = "good"
                status.append(x_stat)
            if mito_y >= roi_y_min and mito_y <= roi_y_max:
                y_stat = "good"
                status.append(y_stat)
            if mito_z >= roi_z_min and mito_z <= roi_z_max:
                z_stat = "good"
                status.append(z_stat)

            if len(status) == 3:
                roi_mitos.append(fix_loc)
                file_points.append({"point loc": fix_loc,"roi name": roi_name})
        print(len(roi_mitos))
        roi_user_data = {"user": user, "roi": roi_name, "mitos in roi": len(roi_mitos)}
        out_data.append(roi_user_data)
    df_points = pd.DataFrame(file_points)
    df_points.to_csv(user+"_file_points_roi_repeats.csv")
    
    merge = df.merge(df_points, how='outer', left_on='location', right_on='point loc')
    print(merge.head)
    merge.to_csv(user+"_file_points_conncat_repeatROIs_09-11-2020.csv")
    print(out_data)
    return(out_data)
```

```python
# files = glob.glob("user_files/*.csv")
# files = glob.glob("repeatROIs/*.csv")
files = glob.glob("prec_rec/*.csv")
data_list = []
```

```python
for file in files:
    print(file)
    if "points" in file:
        type = "point"
        data = get_stuff(file)
        print("data: ", len(data))
        for e in data:
            data_list.append(e)
        print(data_list)
    else:
        type = "sphere"
    print(type)
```

```python
out_df = pd.DataFrame(data_list)
out_df.to_csv("check_points_rois_training_09-11-2020.csv")
```
